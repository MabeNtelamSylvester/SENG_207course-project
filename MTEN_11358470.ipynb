```python
print(redwine['fixed acidity'].mean())
print(redwine['volatile acidity'].mean())
print(redwine['chlorides'].mean())
print(redwine['free sulfur dioxide'].mean())
print(redwine['density'].mean())
print(redwine['quality'].mean())
```

```python
print(redwine['fixed acidity'].mode())
print(redwine['volatile acidity'].mode())
print(redwine['chlorides'].mode())
print(redwine['free sulfur dioxide'].mode())
print(redwine['density'].mode())
print(redwine['quality'].mode())
```

```python
print(redwine['fixed acidity'].median())
print(redwine['volatile acidity'].median())
print(redwine['chlorides'].median())
print(redwine['free sulfur dioxide'].median())
print(redwine['density'].median())
print(redwine['quality'].median())
```

```python
redwine['fixed acidity']=redwine['fixed acidity'].fillna(redwine['fixed acidity'].mode())
redwine['volatile acidity']=redwine['volatile acidity'].fillna(redwine['volatile acidity'].mode())
redwine['chlorides']=redwine['chlorides'].fillna(redwine['chlorides'].median())
redwine['free sulfur dioxide']=redwine['free sulfur dioxide'].fillna(redwine['free sulfur dioxide'].mode())
redwine['density']=redwine['density'].fillna(redwine['density'].median())
redwine['quality']=redwine['quality'].fillna(redwine['quality'].mean())
```

```python
plt.scatter(redwine['alcohol'],redwine['density'])
plt.title('A SCATTER PLOT OF DENSITY AGAINST ALCOHOL')
plt.xlabel('ALCOHOL')
plt.ylabel('DENSITY')
plt.show()
```

* #### A SCATTER PLOT OF DENSITY AGAINST ALCOHOL
  A negative correlation is gotten as alcohol content increases, density decreases.
  
  This reflects the relationship between mass of alcohol and overall density.

```python
plt.bar(redwine['alcohol'],redwine['quality'])
plt.title('A BAR CHART OF QUALITY AGAINST ALCOHOL')
plt.xlabel('ALCOHOL')
plt.ylabel('QUALITY')
plt.show()
```

* #### A BAR CHART OF QUALITY AGAINST ALCOHOL
  A positive correlation is gotten.

  Higher alcohol content may be associated with higher quality rating, as alcohol contributes to the wine's body and mouthfeel. 

```python
plt.bar(redwine['fixed acidity'],redwine['volatile acidity'])
plt.title('A BAR CHART OF VOLATILE ACIDITY AGAINST FIXED ACIDITY')
plt.xlabel('FIXED ACIDITY')
plt.ylabel('VOLATILE ACIDITY')
plt.show()
```

* #### A BAR CHART OF VOLATILE ACIDITY AGAINST FIXED ACIDITY
  A negative correlation is gotten, as fixed acidity increases, volatile acidity decreases and vise versa.

  This correlation provide insight into the balance of acids in the wine.
